```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def f(x):
    return x**2 - 6*x + 8

def bisection(a,b,tol):
    xl = a
    xu = b
    
    i = 1
    while(np.abs(xl-xu)>= tol):
        xr = (xl+xu)/2
        prod = f(xl)*f(xr)
        if prod > tol:
            xl = xr
            i+=1
        else:
            if prod < tol:
                xu = xr
                i+=1
    return i,xr

jawaban = bisection(0.1,0.2,0.01)
print("akar nilai x=", jawaban[1])
print("Iterasi ke-" ,jawaban[0])
x = np.linspace(-2,2,100)


plt.title("metode bisection Penyelesaian persamaan f(x)= x^2 - 6x + 8")
plt.plot(x, f(x))
plt.grid()
plt.show()
```
